```python
from collections import deque
queue = deque(to_mine)
queue
```

```python
def get_user_posts(url_list, influencer_posts_df, username):
    user_post_objects = {}
    for url in tqdm(url_list, total=len(url_list)):
        post_obj = InstagramPost(url)
        user_post_objects[url] = post_obj
        post_data = post_obj.post_data
        post = {}
        post['username'] = username
        post['post_type'] = post_obj.type_of_post
        post['display_url'] = post_obj.display_url
        post['short_url'] = url
        post['is_video'] = post_data['is_video']
        try:
            post['edge_media_to_caption'] = post_data['edge_media_to_caption']['edges'][0]['node']['text']
        except:
            post['edge_media_to_caption'] = ""
        post['edge_media_to_comment'] = post_data['edge_media_to_parent_comment']['count']
        post['comments'] = post_data['edge_media_to_parent_comment']['edges']
        post['taken_at_timestamp'] = post_data['taken_at_timestamp']
        post['edge_liked_by'] = post_obj.number_of_likes
        post['edge_media_to_tagged_users'] = [i['node']['user']['username'] for i in post_data['edge_media_to_tagged_user']['edges']]
        post['video_view_count'] = post_data.get('video_view_count', -1)
        post['video_play_count'] = post_data.get('video_play_count', -1)
        
        # user_posts.append(post)
        influencer_posts_df = influencer_posts_df.append(post, ignore_index=True)
        influencer_posts_df.to_csv('data/influencer_post_db_temp.csv', index=False)
        # print('num posts ', len(influencer_posts_df))
        time.sleep(10)
    return user_post_objects
```

```python
import pickle
def get_user_comments(url_list):
    all_comments_dict = {}
    for url in tqdm(url_list):
        print(url)
        driver.get(url)
        time.sleep(10)
        text_comments_list = []
        soup = BeautifulSoup(driver.page_source,'lxml')
        all_ul = soup.find_all('ul', class_='_a9ym')
        for comment in all_ul:
            temp = comment.find('span', class_='_aacl _aaco _aacu _aacx _aad7 _aade')
            if temp:
                text_comments_list.append(temp.text)
        print(text_comments_list)

            
        all_comments_dict[url] = text_comments_list
        print('all_comments_dict', len(all_comments_dict))
        with open('data/current_all_comments_dict.pkl', 'wb') as g:
            pickle.dump(all_comments_dict, g)
    
    return all_comments_dict
```

```python
# MINE COMMENTS

while queue:
    username = queue.popleft()
    print(f"Mining {username}")
    
    #searchbox
    driver.get(f'https://www.instagram.com/{username}')
    
    #scroll
    scrolldown=driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
    scrolls_left = 2
    influencer_post_url_list = []
    while(scrolls_left > 0):
        scrolldown = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
        scrolls_left -= 1
        
        soup = BeautifulSoup(driver.page_source,'lxml')
        all_url = [f'https://www.instagram.com/{i["href"]}' for i in soup.find_all('a') if i['href'].startswith('/p/')]
        influencer_post_url_list += all_url
        time.sleep(5)
        
    influencer_post_url_list = list(set(influencer_post_url_list))
    print('posts to mine: ', len(influencer_post_url_list))
    
    user_post_objects = get_user_posts(influencer_post_url_list, influencer_posts_df, username)
    all_post_objects[username] = user_post_objects
    
    user_comments_dict = get_user_comments(influencer_post_url_list)
    all_user_comments[username] = user_comments_dict
    
    
    with open('data/influencer_comments.pkl', 'wb') as f:
        pickle.dump(all_user_comments, f)
 
```
